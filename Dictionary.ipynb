# Dictionary

```python
# Function for saving file in text format.

def generate_dictionary_file():
    with open('dictionary.txt', 'w') as file:
        file.write("Word,Meaning\n")

# Function for adding word and meaning and to add the word into dictionary.
def add_word_meaning():
    word = input("Enter the word: ")
    meaning = input("Enter the meaning: ")
    
    with open('dictionary.txt', 'a') as file:
        file.write(f"{word},{meaning}\n")
    
    print(f"Word '{word}' and its meaning '{meaning}' added to the dictionary.")

# Function for searching the word from the dictionary.
def search_word_meaning():
    word_to_search = input("Enter the word to search for: ")
    
    with open('dictionary.txt', 'r') as file:
        lines = file.readlines()
        for line in lines[1:]:
            word, meaning = line.strip().split(',')
            if word == word_to_search:
                print(f"Meaning of '{word_to_search}': {meaning}")
                return
        print(f"'{word_to_search}' not found in the dictionary.")

# Function for removing word from the dictionary.
def remove_word_meaning():
    word_to_remove = input("Enter the word to remove: ")
    
    with open('dictionary.txt', 'r') as file:
        lines = file.readlines()
    
    with open('dictionary.txt', 'w') as file:
        file.write(lines[0])  # Write the header line
        
        removed = False
        for line in lines[1:]:
            word, _ = line.strip().split(',')
            if word == word_to_remove:
                removed = True
            else:
                file.write(line)
        
        if removed:
            print(f"Word '{word_to_remove}' and its meaning removed from the dictionary.")
        else:
            print(f"'{word_to_remove}' not found in the dictionary.")

# Main function for showing the all function in choice.
def main():
    generate_dictionary_file()
    
    while True:
        print("\nDictionary Console Tool")
        print("1. Add Word and Meaning")
        print("2. Search for Meaning")
        print("3. Remove Word and Meaning")
        print("4. Exit")
        
        choice = input("Enter your choice (1/2/3/4): ")
        
        if choice == '1':
            add_word_meaning()
        elif choice == '2':
            search_word_meaning()
        elif choice == '3':
            remove_word_meaning()
        elif choice == '4':
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
